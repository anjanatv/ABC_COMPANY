Segregate employees based on their positions within the company

```python
plt.figure(figsize=(10,6))
abc.plot(kind='bar',color='green',width=0.4,edgecolor='black')
data=position_count
plt.xlabel("Position")
plt.ylabel("count")
plt.title("EMPLOYEES BASED ON THEIR POSTIONS WITHIN THE COMPANY",color='red')
plt.show()
```

Identify the predominant age group among employees. (2 marks)

```python
age_group=abc.value_counts('Current Age').sort_index()
age_group
```

```python
age_bins=[20,25,30,35,40]
abc['age_group']=pd.cut(abc['Current Age'],bins = age_bins)
age_group_distribution=abc['age_group'].value_counts().sort_index()
age_group_distribution_abc=age_group_distribution.reset_index()
age_group_distribution_abc.columns=['age_group','count']
plt.figure(figsize=(8,6))
sns.barplot(x='age_group',y='count',data=age_group_distribution_abc,width=0.5)
plt.title("Predominent Age group of Employees")
plt.xlabel("Current Age")
plt.ylabel("Count")
plt.xticks(35)
abc.rename(columns={"Age":"Current Age"},inplace=True)
plt.show()
```

Discover which team and position have the highest salary expenditure

```python
expenditure=abc.groupby(["Team","Position"])["Salary"].sum().reset_index()
expenditure=expenditure.sort_values(by="Salary", ascending=False)
highest_expenditure=expenditure.iloc[0]
highest_expenditure
```

```python
expenditure_data=expenditure.head(10)
plt.figure(figsize=(10,6))
plt.bar(expenditure_data['Team'],expenditure_data['Salary'],color='red')
plt.title("Salary expenditure by Team and Position")
plt.xlabel("Team","Position")
plt.ylabel("Salary")
plt.show()
```
